```python
import pandas as pd
import numpy as np
```

```python
# Read data and make a copy to manipulate data
df = pd.read_csv('bmw.csv')
```

```python
data = df.copy()
data.shape
```

```python
# Duplicate data represents 1% of the entire dataset so drop them
data.drop_duplicates(inplace=True)
data.shape
```

```python
# Select the most important data
columns = ['model','year','mileage','transmission','']
```

```python
enc_df = pd.get_dummies(data, prefix_sep='_', drop_first=False)
enc_df.head()
```

```python
# Correlation matrix of the numerical features to verify intuition
# The price row shows which features are correlated to the price
enc_df.corr().round(2).style.background_gradient(cmap='coolwarm')
```

```python
# Fuel type, mpg, and tax have little importance so they can be dropped
# mpg and tax are correlated with engine size anyways

cols = ['fuelType','mpg','tax']
final_df = data.drop(columns=cols)
final_df.head()
```

```python
# Perform dummy encoding on the final dataframe and export as csv
final_df_enc = pd.get_dummies(final_df,prefix_sep='_',drop_first=False)
```

```python
final_df_enc.to_csv('car_data_good.csv')
```
