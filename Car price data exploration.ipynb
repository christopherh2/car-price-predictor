```python
# import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
```

```python
# load data
data = pd.read_csv('bmw.csv')
data.head()
```

```python
# Data quality analysis
# Check the shape of the data
data.shape
```

```python
# Summarizing numerical features
data.describe()
```

```python
# Summarizing categorical features
data.describe(include=['object'])
```

```python
# Random 10 samples of the data
data.sample(n=10)
```

```python
# Missing value check
data.isnull().sum()
```

```python
# Duplicate data check
data.duplicated().sum()
```

```python
# Unique value check
data.nunique()
```

```python
# Univariate analysis
sns.histplot(data['model'])
plt.title('Distribution of Models')
plt.xlabel('Model')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.distplot(data['year'])
plt.title('Distribution of Year')
plt.xlabel('Year')
plt.rcParams["figure.figsize"] = (8,3)
plt.show()
```

```python
sns.distplot(data['price'])
plt.xlabel("Price ()")
plt.title("Distribution of Price")
```

```python
sns.histplot(data['transmission'])
plt.title('Distribution of Transmission T?ypes')
plt.xlabel('Transmission')
plt.rcParams["figure.figsize"] = (6,3)
plt.show()
```

```python
sns.distplot(data['mileage'])
plt.xlabel("Mileage (miles)")
plt.title("Distribution of Mileage")
```

```python
sns.distplot(data['tax'])
plt.xlabel("Tax ()")
plt.title("Distribution of Tax")
plt.rcParams["figure.figsize"] = (6,3)
plt.show()
```

```python
sns.distplot(data['mpg'])
plt.xlabel("Engine Efficiency (mpg)")
plt.title("Distribution of MPG")
plt.rcParams["figure.figsize"] = (6,3)
plt.show()
```

```python
# Bivariate analysis
# x vs price is the relationships of interest
data.corr().round(3).style.background_gradient(cmap='coolwarm')
```

```python
sns.boxplot(x=data['model'],y=data['price'])
plt.title('Model vs. Price')
plt.xlabel('Model')
plt.ylabel("Price ()")
plt.xticks(rotation=45)
plt.rcParams["figure.figsize"] = (20,3)
plt.show()
```

```python
sns.boxplot(x=data['year'],y=data['price'])
plt.title('Year vs. Price')
plt.xlabel('Model')
plt.xticks(rotation=45)
plt.ylabel("Price ()")
plt.rcParams["figure.figsize"] = (16,3)
plt.show()
```
